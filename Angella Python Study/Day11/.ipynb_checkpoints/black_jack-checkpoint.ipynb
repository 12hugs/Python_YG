```python
import random
from art import logo

def deal_card():
    """덱에서 무작위 카드를 반환합니다."""
    cards = [11, 2, 3, 4, 5, 6, 7, 8, 9, 10, 10, 10, 10]
    card = random.choice(cards)
    return card

def calculate_score(cards):
    """카드 리스트를 받아서 점수를 계산하여 반환합니다."""
    
    if sum(cards) == 21 and len(cards) == 2:
        return 0
    
    if 11 in cards and sum(cards) > 21:
        cards.remove(11)
        cards.append(1)
    return sum(cards)

def compare(user_score, computer_score):
    if user_score > 21 and computer_score > 21:
        return "둘 다 21을 초과했습니다. 비겼습니다. 😤"

    if user_score == computer_score:
        return "무승부 🙃"
    elif computer_score == 0:
        return "패배, 상대가 블랙잭입니다. 😱"
    elif user_score == 0:
        return "블랙잭으로 승리합니다. 😎"
    elif user_score > 21:
        return "21을 초과했습니다. 패배입니다. 😭"
    elif computer_score > 21:
        return "상대가 21을 초과했습니다. 승리합니다. 😁"
    elif user_score > computer_score:
        return "승리합니다. 😃"
    else:
        return "패배합니다. 😤"

def play_game():

    print(logo)

    user_cards = []
    computer_cards = []
    is_game_over = False

    for _ in range(2):
        user_cards.append(deal_card())
        computer_cards.append(deal_card())

    while not is_game_over:
        user_score = calculate_score(user_cards)
        computer_score = calculate_score(computer_cards)
        print(f"   당신의 카드: {user_cards}, 현재 점수: {user_score}")
        print(f"   컴퓨터의 첫 번째 카드: {computer_cards[0]}")

        if user_score == 0 or computer_score == 0 or user_score > 21:
            is_game_over = True
        else:
            user_should_deal = input("카드를 더 받으려면 'y', 패스하려면 'n'을 입력하세요: ")
            if user_should_deal == "y":
                user_cards.append(deal_card())
            else:
                is_game_over = True

    while computer_score != 0 and computer_score < 17:
        computer_cards.append(deal_card())
        computer_score = calculate_score(computer_cards)

    print(f"   당신의 최종 카드: {user_cards}, 최종 점수: {user_score}")
    print(f"   컴퓨터의 최종 카드: {computer_cards}, 최종 점수: {computer_score}")
    print(compare(user_score, computer_score))

while input("블랙잭 게임을 하시겠습니까? 'y' 또는 'n' 입력: ") == "y":
    play_game()
```
