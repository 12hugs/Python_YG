```python
def is_leap(year):
  if year % 4 == 0: # ëª¨ë“  ë…„ë„ë¥¼ 4ë¡œ ë‚˜ëˆ´ì„ ë•Œ 0ì´ ëœë‹¤ë©´ "ìœ¤ë…„"ì´ ëœë‹¤.
    if year % 100 == 0:
      if year % 400 == 0:
        return True
      else:
        return False
    else:
        return True
  else:
    return False
    
def days_in_month(year,month):
    month_days = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]  
    
    if is_leap(year) and month == 2 :
        return 29
    return month_days[month-1]
        
  
  
#ðŸš¨ ì•„ëž˜ ì½”ë“œëŠ” ì ˆëŒ€ ë³€ê²½í•˜ì§€ ë§ˆì„¸ìš” 
year = int(input("ë…„ë„ë¥¼ ìž…ë ¥í•˜ì„¸ìš”: "))
month = int(input("ì›”ì„ ìž…ë ¥í•˜ì„¸ìš”: "))
days = days_in_month(year, month)
print(days)
```
